```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib
from scipy import stats
import scipy
```

```python
pixW = 64
pixH = 64
```

```python
image = Image.open('sat_at_2611_5858_14naipzoom.png')
arr = np.array(image)
plt.imshow(arr)
```

```python
roadsimage = Image.open('sat_at_2611_5858_14roads.png')
roadsimage.thumbnail((512,512))
roads = np.array(roadsimage)
print(max(roads.flat))
roads = roads.astype('float') / max(roads.flat) 
plt.imshow(roads)
```

```python
redux = (arr // 16) * 16
plt.imshow(redux)
```

```python
pixels = np.zeros((pixH,pixW,3), dtype=np.uint8)
```

```python
wDownscale = len(redux[0]) // pixW
hDownscale = len(redux) // pixH
```

```python
for y in range(pixH - 1):
    for x in range(pixW - 1):
        zone = redux[y*hDownscale:(y+2)*hDownscale, x*wDownscale:(x+2)*wDownscale]
        zone = zone.reshape((hDownscale * wDownscale * 4, 3))
        pixels[y,x] = stats.mode(zone, keepdims=True)[0]
```

```python
plt.imshow(pixels, interpolation="nearest")
```

```python
average_road_color = (arr[roads > 0].sum(axis=0) / len(arr[roads>0])).astype(np.uint8)
withroads = arr.copy()
withroads[roads > 0] = [150,150,150]#average_road_color
```

```python
plt.imshow(withroads,interpolation="nearest")
```

```python
def pixelate(a):
    pixels = np.zeros((pixH,pixW,3), dtype=np.uint8)
    for y in range(pixH ):
        for x in range(pixW):
            zone = a[y*hDownscale:(y+1)*hDownscale, x*wDownscale:(x+1)*wDownscale]
            zone = zone.reshape((hDownscale * wDownscale, 3))
            # pixels[y,x] = stats.mode(zone, keepdims=True)[0]
            # pixels[y,x] = zone[0]
            pixels[y,x] = np.mean(zone, axis=0)
    return pixels
```

```python
plt.imshow(pixelate(arr), interpolation="nearest")
```

```python
plt.imshow(pixelate(withroads), interpolation="nearest")
```
